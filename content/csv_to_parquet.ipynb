```python
import pyarrow.csv as pv
import pyarrow.parquet as pq

# Read and write the table with pyarrow
table = pv.read_csv('2022_12.csv')
pq.write_table(table, '2022_12.csv'.replace('csv', 'parquet'))
```

```python
# Once converted to store, the below can be used to read the new parquet file

import pandas as pd

# Read the parquet file
pd.read_parquet('MyLA311_Service_Request_Data_2022.parquet', engine='pyarrow')
```
